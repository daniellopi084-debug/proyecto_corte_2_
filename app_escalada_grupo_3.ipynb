### Descripción del Proyecto
Este sistema representa una reingeniería completa del MVP inicial, migrando de una lógica simple a una arquitectura profesional basada en:
1. **POO:** Encapsulamiento, Herencia y Polimorfismo.
2. **Persistencia:** Base de datos SQLite normalizada (3NF).
3. **Robustez:** Manejo de excepciones (Try-Except) y validaciones.

```python
import sqlite3

# --- ESTRUCTURA DE CLASES (REQUISITO POO) ---
class Entidad:
    def __init__(self, nombre):
        self._nombre = nombre # Encapsulamiento protegido

class Producto(Entidad):
    def __init__(self, nombre, precio, stock, nit_prov):
        super().__init__(nombre)
        self.__precio = precio # Atributo privado
        self.stock = stock
        self.nit_prov = nit_prov
    def get_precio(self): return self.__precio

class Persona:
    def __init__(self, id_per, nombre):
        self.id_per = id_per
        self.nombre = nombre

class Proveedor(Persona):
    def __init__(self, nit, nombre, ciudad):
        super().__init__(nit, nombre)
        self.ciudad = ciudad

class Operacion:
    def __init__(self, tipo):
        self.tipo = tipo

class Venta(Operacion):
    def __init__(self, tipo, id_prod, cant):
        super().__init__(tipo)
        self.id_prod = id_prod
        self.cant = cant

# --- SISTEMA DE PERSISTENCIA (SQLITE) ---
class SistemaDaniel:
    def __init__(self):
        try:
            # Nombre de BD según instrucción del profesor
            self.conexion = sqlite3.connect("base_datos_startup_grupo_final.db")
            self.cursor = self.conexion.cursor()
            self._configurar_bd()
        except sqlite3.Error as e:
            print(f"Error: {e}")

    def _configurar_bd(self):
        # 1. Crear 3 tablas relacionadas
        self.cursor.execute("CREATE TABLE IF NOT EXISTS proveedores (nit TEXT PRIMARY KEY, nom TEXT, ciudad TEXT)")
        self.cursor.execute("CREATE TABLE IF NOT EXISTS productos (id INTEGER PRIMARY KEY AUTOINCREMENT, nom TEXT, precio REAL, stock INTEGER, nit_p TEXT, FOREIGN KEY(nit_p) REFERENCES proveedores(nit))")
        self.cursor.execute("CREATE TABLE IF NOT EXISTS ventas (id_v INTEGER PRIMARY KEY AUTOINCREMENT, id_p INTEGER, cant INTEGER, FOREIGN KEY(id_p) REFERENCES productos(id))")
        
        # 2. Insertar 5 registros (Regla de Oro)
        p_datos = [('101', 'Café Central', 'Chía'), ('102', 'Panadería Sol', 'Bogotá'), ('103', 'Lácteos Up', 'Sopó'), ('104', 'Frutas S.A', 'Cajicá'), ('105', 'Envases IQ', 'Chía')]
        self.cursor.executemany("INSERT OR IGNORE INTO proveedores VALUES (?,?,?)", p_datos)
        
        prod_datos = [('Tinto', 2500, 50, '101'), ('Croissant', 3500, 20, '102'), ('Avena', 4000, 15, '103'), ('Jugo Naranja', 5000, 10, '104'), ('Vaso Té', 3000, 30, '105')]
        self.cursor.executemany("INSERT OR IGNORE INTO productos (nom, precio, stock, nit_p) VALUES (?,?,?,?)", prod_datos)
        self.conexion.commit()

    def generar_reporte(self):
        # 3. Consulta con JOIN (Requisito SQL)
        print("\n" + "="*40)
        print("SISTEMA DE CAFETERÍA - REPORTE PROFESIONAL")
        print("="*40)
        self.cursor.execute("SELECT p.nom, p.precio, pr.nom FROM productos p JOIN proveedores pr ON p.nit_p = pr.nit")
        for fila in self.cursor.fetchall():
            print(f"PRODUCTO: {fila[0]} | PRECIO: ${fila[1]} | PROVEEDOR: {fila[2]}")
        print("="*40)

# --- EJECUCIÓN DIRECTA ---
if __name__ == "__main__":
    # Inicia el sistema sin detenerse a pedir inputs
    app = SistemaDaniel()
    app.generar_reporte()
    print("\n✅ Base de datos generada y 5 registros cargados con éxito.")
```
